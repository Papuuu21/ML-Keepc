## **COMPARACIÓN CON STATSMODELS**

    Usamos el preprocesado inicial de partida:

```python
import numpy  as np  
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
from utils import *

# Cargamos sólo los datos de train:
preprocess_train = pd.read_csv('./prepoces_day_train.csv', sep=';', decimal='.')
preprocess_test = pd.read_csv('./preproces_day_test.csv', sep=';', decimal='.')
```

```python
y_train = preprocess_train['Price']
x_train = preprocess_train.drop(columns = 'Price')
```

```python
X = sm.add_constant(x_train)
results = sm.OLS(y_train, X).fit()
y_train_pred = results.predict(X)
results.summary()
```

```python
# Calcular el MSE
mse_train = results.ssr / results.nobs
print(f"MSE train: {mse_train}")
# Calcular el RMSE
rmse_train = np.sqrt(np.mean((y_train - y_train_pred)**2))
print(f"RMSE: {rmse_train}")
```

```python
# SELECCIÓN DE VARIABLES BASADAS EN P-VALOR:
p_valor_alto = ['Host Response Rate', 'Security Deposit', 'Guests Included', 'Minimum Nights', 'Maximum Nights', 'Calculated host listings count', 'Bed Type codif']

x_train2 = preprocess_train.drop(columns = 'Price')
x_train2 = preprocess_train.drop(columns = p_valor_alto)
```

```python
X2 = sm.add_constant(x_train2)
results = sm.OLS(y_train, X2).fit()
y_pred2 = results.predict(X2)
results.summary()
```

```python
# Calcular el MSE
mse2_train = results.ssr / results.nobs
print(f"MSE2 train: {mse2_train}")
# Calcular el RMSE
rmse2_train = np.sqrt(np.mean((y_train - y_pred2)**2))
print(f"RMSE2 train: {rmse2_train}")
```

```python
y_test = preprocess_test['Price']
x_test = preprocess_test.drop(columns = 'Price')
```

```python
# SELECCIÓN DE VARIABLES BASADAS EN P-VALOR:
p_valor_alto = ['Host Response Rate', 'Security Deposit', 'Guests Included', 'Minimum Nights', 'Maximum Nights', 'Calculated host listings count', 'Bed Type codif']

x_test2 = preprocess_test.drop(columns = 'Price')
x_test2 = preprocess_test.drop(columns = p_valor_alto)
```

```python
X_test = sm.add_constant(x_test2)
results = sm.OLS(y_test, x_test2).fit()
y_pred_test = results.predict(x_test2)
results.summary()
```
